# Data visualization and analysis





```python
# visualization
import pandas as pd
from plotly.graph_objs import Trace
import matplotlib.pyplot as plt
import plotly.express as px
```

**import data**

```python
df=pd.read_csv('amazon_data_science_books_modified.csv')
df
```

**Scatter plot**

```python

fig1=px.scatter(df , x='price', y='pages',color='n_reviews',size='avg_reviews')
fig1.update_layout(
    width = 1200,
    height = 800,
    title = "Price, Number of Pages, Average Reviews and n_reviews")
fig1.show()
```

```python
df['n_reviews'].replace(',','.', regex=True).astype(float).sort_index()
df['n_reviews'] = pd.to_numeric(df['n_reviews'], errors='coerce')
fifteen_most_reviewed_books = df.nlargest(15, 'n_reviews' )     # 15 of the largest 'n_reviews'

fifteen_most_reviewed_books
```

**Plotly bar plots**

```python

fig=px.bar(fifteen_most_reviewed_books,x='title',y='avg_reviews',log_y=True)
fig.update_layout(
    width = 800,
    height = 1500,
    title = "fifteen Most Reviewed Books: Average Reviews")
fig.show()
```

```python
fig=px.bar(fifteen_most_reviewed_books, x= 'title' , y=['star5','star4','star3','star2','star1'],log_y=True)
fig.update_layout(
    width = 800,
    height = 900,
    title = "Fifteen Most Expensive Books: Number of Stars Percentage")
fig.show()
```
