```python

# data analysis and wrangling

from pandas.core import apply
from numpy import size
from pickle import FLOAT
import pandas as pd

# visualization

from plotly.graph_objs import Trace
import matplotlib.pyplot as plt
import plotly.express as px

```

**import data**

```python
df=pd.read_csv('Amazon Data Science Books.csv')
df
```

```python
# check #na columns

df.isna().sum()
```

***start Data cleaning***

```python
#Remove nan values

df.dropna(inplace=True)
df
```

```python
# Remove the bracket from the author column

def handle_author (author):
  return author.replace("[","").replace("]","")

df['author']=df.author.apply(lambda author:handle_author(author))
df['author']=df.author.astype('string')   # Change format to string
df
```

```python
# Change weight to kilograms

def weight_convertor(item):

  if item.split()[-1] == "pounds":
      return float (item.split()[0]) * 0.45359237

  elif item.split()[-1] =="ounces":
      return float (item.split()[0]) * 0.02834952

df["weight"]=df["weight"].apply(weight_convertor)
```

```python
#Cut the useful part from the publisher

df['publisher']=df['publisher'].apply(lambda x : x.split(';')[0])
df['publisher']=df.publisher.astype('string')   # Change format to string
df
```

```python
# Convert dimensions to cm

def dimensions_conv (item):

  if item.split()[:-1]. __len__() == 5 :
     x = float(item.split()[:-1][0]) * 2.54  #cm
     y = float(item.split()[:-1][2]) * 2.54  #cm
     z = float(item.split()[:-1][4]) * 2.54  #cm

     return round(x*y*z ,1)

  else:
    return None

df['size'] = df['dimensions'].apply(dimensions_conv)
df
```

```python
# Moving column size to 13 columns

column_to_move = df.pop("size")
df.insert(13, "size", column_to_move)
```

```python
# Remove the Spanish column

df.drop(df.loc[df['language']=='Spanish'].index, inplace=True)
df
```

**best top10 ML books**

```python
# Select books based on title containing "Python"
df['title']=df.title.astype('string')
ML_books = df[df['title'].str.contains("Machine Learning")]

# Python books with most reviews and highest average rating
best_ML_books = ML_books.nlargest(10, ['avg_reviews'])
best_ML_books
```
